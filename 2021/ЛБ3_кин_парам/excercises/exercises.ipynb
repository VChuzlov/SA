```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
import pandas as pd
```

```python
def main(foo, c, k, time, cols):
    res = solve_ivp(foo, (time[0], time[-1]), c, args=(k, ), t_eval=time)
    
    plt.style.use('ggplot')
    plt.plot(time, res.y[0], label=cols[0])
    plt.plot(time, res.y[1], label=cols[1])
    plt.plot(time, res.y[2], label=cols[2])
    plt.plot(time, res.y[3], label=cols[3])
    plt.ylabel('Components concentration, $mole / cm^3$')
    plt.xlabel('Reaction time, sec.')
    plt.legend()
    plt.tight_layout()
    
    df = pd.DataFrame(res.y.T, columns=cols, index=time)
    
    return df
```

## Var 1

```python
def f1(time, c, k):
    right_parts = np.zeros(c.shape[0])
    right_parts[0] = -k[0] * c[0]
    right_parts[1] = k[0] * c[0] - k[1] * c[1]
    right_parts[2] = k[1] * c[1]
    right_parts[3] = k[0] * c[0] + k[1] * c[1]
    
    return right_parts
```

```python
c = np.array([1, 0, 0, 0])
k = np.array([1.25, 0.85])
time = np.arange(0, 1.1, 0.1)
cols=['C10H22', 'C10H20', 'C10H18', 'H2']
```

```python
df = main(f1, c, k, time, cols)
```

```python
df
```
