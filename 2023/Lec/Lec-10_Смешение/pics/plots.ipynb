```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator
```

```python
data = np.array(
    [
        0.0073, 0.2374,  .0,    .0, 0.0103, 0.2064, 0.031,  .0,
        0.1032, 0.1961, 0.1239, .0, 0.0062, 0.0361, 0.0206, .0, 0.0215,
    ]
)
```

```python
f, a = plt.subplots(figsize=(5.5, 5))

a.scatter(data, calc, alpha=.5)

max_value = np.max((data.max(), calc.max()))
line = np.linspace(0, max_value, 5)
a.plot(line, line, '--k', alpha=.5)

a.set_ylabel('Расчетное значение концентрации компонента')
a.set_xlabel('Экспериментальное значение концентрации компонента')

a.xaxis.set_minor_locator(AutoMinorLocator())
a.yaxis.set_minor_locator(AutoMinorLocator())

plt.tight_layout()
plt.savefig('err.pdf')
```
