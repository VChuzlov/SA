```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from matplotlib.ticker import AutoMinorLocator
```

```python
delta = 0.025
value = 5
x = np.arange(-value, value+delta, delta)
y = np.arange(-value, value+delta, delta)
X, Y = np.meshgrid(x, y)
Z = (X ** 2 + Y - 11) ** 2 + (X + Y ** 2 - 7) ** 2
```

```python
fig, ax = plt.subplots()
cs = ax.contourf(X, Y, Z, 200, cmap='jet')
# ax.clabel(cs, inline=True, fontsize=10)
ax.xaxis.set_minor_locator(AutoMinorLocator())
ax.yaxis.set_minor_locator(AutoMinorLocator())
ax.set_xlabel('X')
ax.set_ylabel('Y')
fig.colorbar(cs)
plt.tight_layout()
plt.savefig('contour.pdf')
```
