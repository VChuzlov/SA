```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from matplotlib.ticker import AutoMinorLocator
```

```python
delta = 0.025
value = 5
x = np.arange(-value, value+delta, delta)
y = np.arange(-value, value+delta, delta)
X, Y = np.meshgrid(x, y)
Z = (X ** 2 + Y - 11) ** 2 + (X + Y ** 2 - 7) ** 2
```

```python
fig, ax = plt.subplots()
vmin, vmax = 0, 182
cs = ax.contourf(
    X, Y, Z, 
    cmap='jet',
    levels=np.linspace(vmin, vmax, 150),
    extend='niether'
)
cs.cmap.set_under('white')
# ax.clabel(cs, inline=True, fontsize=10)
ax.xaxis.set_minor_locator(AutoMinorLocator())
ax.yaxis.set_minor_locator(AutoMinorLocator())
ax.set_xlabel('X')
ax.set_ylabel('Y')
fig.colorbar(
    cm.ScalarMappable(
        norm=cs.norm, 
        cmap=cs.cmap
    ),
    ticks=range(vmin, vmax+10, 10)
)
plt.tight_layout()
plt.savefig(r'.\pics\contour2.pdf')
```
