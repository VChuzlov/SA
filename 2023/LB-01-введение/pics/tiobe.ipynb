```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.style.use('fivethirtyeight')
```

```python
plt.rcdefaults()
fig, ax = plt.subplots()

# Example data
langs = ('Python', 'C', 'C++', 'Java', 'C#', 'Visual Basic', 'JavaScript', 'SQL', 'Assembly Language', 'PHP')
y_pos = np.arange(len(langs))
performance = (16.36, 16.26, 12.91, 12.21, 5.73, 4.64, 2.87, 2.50, 1.60, 1.39)

ax.barh(y_pos, performance, align='center', color=(40/255, 190/255, 70/255))
ax.set_yticks(y_pos)
ax.set_yticklabels(langs)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Рейтинг, %')

plt.tight_layout()
plt.savefig('tiobe.pdf')
```
