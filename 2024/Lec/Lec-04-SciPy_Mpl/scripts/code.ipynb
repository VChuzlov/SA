```python
import numpy as np
import scipy.optimize as opt


def f(x: float) -> float:
    return 1 / 5 + x * np.cos(3 / x)


root = opt.brentq(f, -.7, -.5)
print(root)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def fun(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

x_mesh, y_mesh = np.meshgrid(x, y)
z_mesh = fun(x_mesh, y_mesh)
	
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(x_mesh, y_mesh, z_mesh, 50, cmap='cool')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.show()
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(x_mesh, y_mesh, z_mesh, 50, cmap='cool')
ax.set_xlabel('x')
ax.set_ylabel('y') 
ax.set_zlabel('z')
ax.view_init(60, 35)
plt.show()
```
